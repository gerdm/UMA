```python
import numpy as np
import sys
sys.path.append("..")
from ido import Simplex
```

# Big M: 1

```python
M = 10000; c = 3

A = np.array([[1, 1, 1, 1, 0,  0, 0],
              [2, 1, 0, 0, 1,  0, 0],
              [2, 1, 3, 0, 0, -1, 1]])

b = np.array([[2, 3, 3]])

z_coef = np.array([[2*M-4, M-4, 3*M-1, 0, 0,-M, 0]])

pr1 = Simplex(A, b, z_coef)
pr1.BigM(M, c)
```

# Big M: 2
