```python
import numpy as np
from numpy.linalg import inv
import matplotlib.pyplot as plt
%matplotlib notebook
```

```python
s = np.array([10, 40, 80, 140, 200, 300])
y = np.array([3.1, 11.9, 21, 29.9, 37.3, 42.7])
```

```python
a, b = 40, 5e-3
```

```python
gi = lambda a, b, s, y: a * (1 - np.exp(-b * s)) - y
g = np.array([[gi(a, b, si, yi) for (si, yi) in zip(s, y)]]).T
```

```python
plt.scatter(s, y);
# First Guess
plt.plot(s, a * (1 - np.exp(-b*s)))
# After First Iteration
plt.plot(s,  4.60813365e+01 * (1 - np.exp(-7.61743682e-03*s)))
```

```python
Jg_0 = 1 - np.exp(-s * b)
Jg_1 = s * a * np.exp(-s*b)
Jg = np.array([Jg_0, Jg_1]).T
Jg
```

```python
np.array([[40, 0.005]]).T - np.dot(inv(np.dot(Jg.T, Jg)), np.dot(Jg.T, g))
```
